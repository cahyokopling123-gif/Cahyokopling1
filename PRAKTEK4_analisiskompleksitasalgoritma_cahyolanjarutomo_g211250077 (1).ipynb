```python
#Linear Search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
# contoh
data = [10, 20, 30, 40, 50]
print(linear_search(data, 30))  # output: 2
```

**Linear Search**

Algoritma ini memeriksa setiap elemen dalam array secara berurutan dari indeks awal hingga akhir.  

1. Best Case: $O(1)$ — Terjadi jika elemen yang dicari (target) berada
di indeks pertama array.

2. Average Case: $O(n)$ — Terjadi jika target ditemukan di suatu tempat di tengah array. Algoritma rata-rata perlu memeriksa setengah dari total elemen.

3. Worst Case: $O(n)$ — Terjadi jika target berada di elemen paling akhir array atau tidak ditemukan sama sekali, sehingga algoritma harus memeriksa seluruh $n$ elemen.

```python
# Binary Search (Iteratif)
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
# contoh (data harus terurut)
data = [10, 20, 30, 40, 50]
print(binary_search(data, 40))  # output:3
```

**Binary Search (Iteratif)**
1. Best Case: $O(1)$ — Terjadi jika target langsung ditemukan di elemen tengah (mid) pada pengecekan pertama.

2. Average Case: $O(\log n)$ — Terjadi jika target ditemukan setelah beberapa kali pembagian rentang pencarian.

3. Worst Case: $O(\log n)$ — Terjadi jika target berada di ujung array ekstrim atau tidak ada dalam array, sehingga proses pembelahan dilakukan sampai rentang pencarian habis.

```python
# Binary Search (Rekursif)
def binary_search_recursive(arr, left, right, target):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_recursive(arr, mid + 1, right, target)
    else:
        return binary_search_recursive(arr, left, mid - 1, target)
# contoh
data = [10, 20, 30, 40, 50]
print(binary_search_recursive(data, 0, len(data)-1, 20))  # output: 1
```

**Binary Search (Rekursif)**
1. Best Case: $O(1)$ — Terjadi jika target langsung ditemukan di elemen tengah (mid) pada pengecekan pertama.

2. Average Case: $O(\log n)$ — Terjadi jika target ditemukan setelah beberapa kali pembagian rentang pencarian.

3. Worst Case: $O(\log n)$ — Terjadi jika target berada di ujung array ekstrim atau tidak ada dalam array, sehingga proses pembelahan dilakukan sampai rentang pencarian habis.

```python
# Jump Search
import math
def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0
    # lompat blok
    while arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1
    # linear search di blok
    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i
    return -1
# contoh
data = [10, 20, 30, 40, 50, 60, 70]
print(jump_search(data, 60))  # output: 5
```

**Jump Search**

Algoritma ini mengoptimalkan pencarian dengan melompati beberapa elemen sekaligus (biasanya sebesar blok akar kuadrat dari jumlah elemen, $\sqrt{n}$), kemudian melakukan pencarian linear (Linear Search) di dalam blok tempat elemen tersebut diperkirakan berada.

1. Best Case: $O(1)$ — Terjadi jika target berada di elemen pertama pada blok pertama yang diperiksa.

2. Average Case: $O(\sqrt{n})$ — Terjadi jika target berada di suatu blok di tengah array dan di tengah pencarian linear pada blok tersebut.

3. Worst Case: $O(\sqrt{n})$ — Terjadi jika target ada di elemen terakhir dari array (sehingga harus melompat sebanyak $\sqrt{n}$ kali) dan elemen tersebut ada di akhir blok saat dilakukan pencarian linear, atau jika elemen tidak ditemukan sama sekali.

```python
# Interpolation Search
def interpolation_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high and target >= arr[low] and target <= arr[high]:
        if low == high:
            if arr[low] == target:
                return low
            return -1
        # rumus interpolasi
        pos = low + ((target - arr[low]) * (high - low) //
                     (arr[high] - arr[low]))
        if arr[pos] == target:
            return pos
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1
    return -1
# contoh
data = [10, 20, 30, 40, 50]
print(interpolation_search(data, 50))  # output: 4
```

**Interpolation Search**

Algoritma ini adalah bentuk pengembangan dari Binary Search. Alih-alih selalu mengecek nilai tengah, Interpolation Search akan "menebak" posisi target berdasarkan perbandingan jarak nilai target dengan nilai terkecil dan terbesar pada array.  

1. Best Case: $O(1)$ — Terjadi jika perhitungan rumus interpolasi pertama langsung menunjuk tepat ke indeks target.

2. Average Case: $O(\log (\log n))$ — Terjadi jika nilai-nilai dalam array terdistribusi secara seragam (jarak antar nilainya konsisten dan merata). Algoritma ini bisa menemukan target sangat cepat karena tebakannya presisi.

3. Worst Case: $O(n)$ — Terjadi jika data melonjak secara ekstrim atau eksponensial (tidak terdistribusi merata). Hal ini menyebabkan rumus tebakan meleset jauh di setiap iterasi, sehingga performanya turun drastis layaknya Linear Search.
