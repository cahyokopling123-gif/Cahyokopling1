Fungsi Dijkstra(Graph, Start):
    1. Buat struktur data `distances` (jarak) dan atur jarak ke semua node menjadi TAK_TERHINGGA (Infinity), kecuali Start = 0.
    2. Buat `priority_queue` untuk menyimpan node yang akan dieksplorasi beserta jaraknya. Masukkan (0, Start).
    3. Buat kumpulan (set) `visited` untuk mencatat node yang sudah diproses secara final.

    4. Selama `priority_queue` tidak kosong:
        a. Ambil `current_node` dengan jarak terkecil dari `priority_queue`.
        b. Jika `current_node` sudah ada di `visited`, lewati dan lanjut ke iterasi berikutnya.
        c. Tambahkan `current_node` ke `visited`.
        d. Untuk setiap `neighbor` (tetangga) dari `current_node`:
            i. Hitung `jarak_baru` = distances[current_node] + bobot jarak ke neighbor.
            ii. Jika `jarak_baru` < distances[neighbor]:
                - Perbarui distances[neighbor] = jarak_baru.
                - Masukkan (jarak_baru, neighbor) ke `priority_queue`.

    5. Kembalikan `distances`

```python
import heapq

def shortest_path_dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances

graph_peta = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 1, 'D': 5},
    'C': {'A': 2, 'B': 1, 'D': 8, 'E': 10},
    'D': {'B': 5, 'C': 8, 'E': 2},
    'E': {'C': 10, 'D': 2}
}

titik_awal = 'A'
hasil_jarak = shortest_path_dijkstra(graph_peta, titik_awal)

print(f"Jarak terpendek dari titik {titik_awal}:")
for node, jarak in hasil_jarak.items():
    print(f"Ke {node} = {jarak}")
```

3**. Analisa Kompleksitas Waktu**

Berdasarkan implementasi di atas (menggunakan Adjacency List dan Min-Priority Queue/Binary Heap):

Kompleksitas Waktu (Time Complexity): O((V + E)log V)
- V merepresentasikan jumlah Vertex (simpul/titik)
- E merepresentasikan jumlah Edges (sisi/garis penghubung).

**PROSES ANALISI**
1. Proses mengeluarkan node dengan nilai minimum dari priority queue membutuhkan waktu O(log V). Karena dalam kasus terburuk kita memasukkan semua node, operasinya menjadi O(V log V).
2. Proses mengevaluasi tetangga (edges relaxation) berjalan sebanyak E kali. Setiap kali kita menemukan jarak baru yang lebih pendek, kita memperbarui/menambahkan data ke dalam priority queue (yang membutuhkan waktu O(log V)). Ini memakan waktu O(E log V).
3. Jika dijumlahkan, total waktunya adalah O(V log V + E log V), yang bisa disederhanakan menjadi O((V + E) log V). Pada graf yang saling terhubung utuh, E umumnya lebih besar dari V, sehingga sering disingkat menjadi O(E log V).
